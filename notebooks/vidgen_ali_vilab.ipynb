```python
import torch
from diffusers import DiffusionPipeline, DPMSolverMultistepScheduler
from diffusers.utils import export_to_video
import os
import numpy as np
from PIL import Image

os.makedirs("out", exist_ok=True)
```

```python
pipe = DiffusionPipeline.from_pretrained("damo-vilab/text-to-video-ms-1.7b", torch_dtype=torch.float16, variant="fp16")
pipe.scheduler = DPMSolverMultistepScheduler.from_config(pipe.scheduler.config)

# optimize for GPU memory
pipe.enable_model_cpu_offload()
pipe.enable_vae_slicing()
```

```python
for scene in prompts['scene_prompts']:
  final_prompt = (
    # prompts['global_prompt']["positive_prompt"]
    # + " "  +
    scene["prompt"]
  )

  # print(final_prompt)

# load pipeline

  # generate
  video_frames = pipe(final_prompt, num_inference_steps=35, num_frames=scene["frames"]).frames

  # convent to video
  video_path = export_to_video(
    video_frames[0],
    output_video_path=f"out/{scene['scene_id']}.mp4",
    fps=12
  )
```

```python
import os
import subprocess

def join_videos_lossless(video_folder, output_filename):
    list_file = "files.txt"

    # Get all .mp4 files and sort them alphabetically
    videos = sorted([f for f in os.listdir(video_folder) if f.endswith('.mp4')])

    # Create a text file containing the list of videos to merge
    with open(list_file, "w", encoding="utf-8") as f:
        for video in videos:
            full_path = os.path.abspath(os.path.join(video_folder, video))
            f.write(f"file '{full_path}'\n")

    # Run the FFmpeg command
    command = [
        "ffmpeg", "-f", "concat", "-safe", "0",
        "-i", list_file, "-c", "copy", output_filename
    ]

    try:
        subprocess.run(command, check=True)
        print(f"Successfully joined videos into {output_filename}")
    except subprocess.CalledProcessError as e:
        print(f"Error during video joining: {e}")
    finally:
        # Clean up temporary list file
        if os.path.exists(list_file):
            os.remove(list_file)

join_videos_lossless("out/", "merged_output.mp4")
```
