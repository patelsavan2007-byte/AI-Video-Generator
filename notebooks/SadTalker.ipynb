```python
import os, re

# All numpy aliases removed in 1.24+
patches = [
    (r'np\.float(?![a-zA-Z0-9_])',   'float'),
    (r'np\.int(?![a-zA-Z0-9_])',     'int'),
    (r'np\.bool(?![a-zA-Z0-9_])',    'bool'),
    (r'np\.complex(?![a-zA-Z0-9_])', 'complex'),
    (r'np\.object(?![a-zA-Z0-9_])',  'object'),
    (r'np\.str(?![a-zA-Z0-9_])',     'str'),
]

patched = []
for root, _, files in os.walk('/content/SadTalker'):
    for fname in files:
        if not fname.endswith('.py'):
            continue
        fpath = os.path.join(root, fname)
        try:
            with open(fpath, 'r', errors='ignore') as f:
                original = f.read()
            updated = original
            for pattern, replacement in patches:
                updated = re.sub(pattern, replacement, updated)
            if updated != original:
                with open(fpath, 'w') as f:
                    f.write(updated)
                patched.append(os.path.relpath(fpath, '/content/SadTalker'))
        except Exception as e:
            print(f"вљ пёЏ  Skipped {fpath}: {e}")

print(f"вњ… Patched {len(patched)} file(s):")
for f in patched: print(f"   {f}")
```

```python
# Patch basicsr to work with torchvision >= 0.16
deg_path = '/usr/local/lib/python3.12/dist-packages/basicsr/data/degradations.py'

with open(deg_path, 'r') as f:
    content = f.read()

content = content.replace(
    'from torchvision.transforms.functional_tensor import rgb_to_grayscale',
    'from torchvision.transforms.functional import rgb_to_grayscale'
)

with open(deg_path, 'w') as f:
    f.write(content)

print("вњ… Patched successfully")
```

```python
preprocess_path = '/content/SadTalker/src/face3d/util/preprocess.py'

with open(preprocess_path, 'r') as f:
    content = f.read()

# Force scalar extraction so np.array([...]) gets 5 plain floats
content = content.replace(
    'trans_params = np.array([w0, h0, s, t[0], t[1]])',
    'trans_params = np.array([float(w0), float(h0), float(s), float(t[0].flat[0]), float(t[1].flat[0])])'
)

with open(preprocess_path, 'w') as f:
    f.write(content)

print("вњ… Patched preprocess.py")
```
